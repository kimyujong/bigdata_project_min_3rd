```python
def fill_missing_values(df, num_features_zero=[], num_features_median=[], cat_features=[]):
    """
    결측 처리:
    - num_features_zero: 0 대체
    - num_features_median: 중앙값 대체
    - cat_features: "Unknown" + missing flag
    """
    df = df.copy()
    
    # 1) 거래금액 등 의미 있는 0
    for col in num_features_zero:
        if col in df.columns:
            df[col] = df[col].fillna(0)
    
    # 2) 연속형 → 중앙값
    for col in num_features_median:
        if col in df.columns:
            median_val = df[col].median()
            df[col] = df[col].fillna(median_val)
    
    # 3) 범주형 → Unknown + flag
    for col in cat_features:
        if col in df.columns:
            df[col + "_missing"] = df[col].isna().astype(int)
            df[col] = df[col].fillna("Unknown")
    
    return df



# 개인CB 파생변수

def create_features_personal(df_09):
    feats = pd.DataFrame(index=df_09.index)
    
    # 소비 패턴
    if "3개월내(15일)카드총이용금액(미해지)" in df_09 and "1년내(15일)카드총이용금액(미해지)" in df_09:
        feats["소비비율_3Mvs1Y"] = df_09["3개월내(15일)카드총이용금액(미해지)"] / (df_09["1년내(15일)카드총이용금액(미해지)"] + 1)
    
    if "3개월내(15일)신용카드할부총이용금액합계(해지포함)" in df_09 and "3개월내(15일)신용카드일시불총이용금액(해지포함)" in df_09:
        feats["할부비율_3M"] = df_09["3개월내(15일)신용카드할부총이용금액합계(해지포함)"] / (
            df_09["3개월내(15일)신용카드할부총이용금액합계(해지포함)"] + df_09["3개월내(15일)신용카드일시불총이용금액(해지포함)"] + 1
        )
    
    # 카드당 평균 사용액
    if "신용카드건수(미해지)" in df_09:
        feats["카드평균이용액_3M"] = df_09["3개월내(15일)카드총이용금액(미해지)"] / (df_09["신용카드건수(미해지)"] + 1)
    
    # 신규카드 비율
    if "신용카드건수(해지포함)(1개월내신규개설)" in df_09:
        feats["신규카드비율_6M"] = (
            df_09["신용카드건수(해지포함)(1개월내신규개설)"].fillna(0) +
            df_09["신용카드건수(해지포함)(3개월내신규개설)"].fillna(0) +
            df_09["신용카드건수(해지포함)(6개월내신규개설)"].fillna(0)
        ) / (df_09["신용카드건수(미해지)"] + 1)
    
    # 체크카드 비율
    if "체크카드건수(미해지)" in df_09:
        feats["체크카드비율"] = df_09["체크카드건수(미해지)"] / (
            df_09["신용카드건수(미해지)"] + df_09["체크카드건수(미해지)"] + 1
        )
    
    # 한도 대비 이용
    if "카드총한도금액(활성카드)(미해지)" in df_09:
        feats["한도소진율"] = df_09["3개월내(15일)카드총이용금액(미해지)"] / (df_09["카드총한도금액(활성카드)(미해지)"] + 1)
        feats["단기대출비율"] = df_09["신용카드총단기카드대출(CA)한도금액(활성카드)(미해지)"] / (df_09["카드총한도금액(활성카드)(미해지)"] + 1)
    
    # 신용이력
    if "최초신용카드개설일자로부터경과일수(활성카드)(미해지)" in df_09:
        feats["최초개설_경과일"] = df_09["최초신용카드개설일자로부터경과일수(활성카드)(미해지)"]
    if "최초카드개설일로부터경과일수(활성카드)(미해지)" in df_09:
        feats["카드개설_경과일"] = df_09["최초카드개설일로부터경과일수(활성카드)(미해지)"]
    
    return feats



# 통신CB 파생변수
def create_features_telco(df_11):
    feats = pd.DataFrame(index=df_11.index)
    
    # 1) 연체율 (기간 대비 보정)
    if "연체건수(1개월내발생)(해제포함)_전년말기준" in df_11:
        feats["연체율_1M"] = df_11["연체건수(1개월내발생)(해제포함)_전년말기준"] / 1
    if "연체건수(3개월내발생)(해제포함)_전년말기준" in df_11:
        feats["연체율_3M"] = df_11["연체건수(3개월내발생)(해제포함)_전년말기준"] / 3
    if "연체건수(6개월내발생)(해제포함)_전년말기준" in df_11:
        feats["연체율_6M"] = df_11["연체건수(6개월내발생)(해제포함)_전년말기준"] / 6
    if "연체건수(1년내발생)(해제포함)_전년말기준" in df_11:
        feats["연체율_1Y"] = df_11["연체건수(1년내발생)(해제포함)_전년말기준"] / 12
    
    # 2) 연체 심각도
    if "최장연체일수_미해제_전년말기준" in df_11:
        feats["최장연체일수"] = df_11["최장연체일수_미해제_전년말기준"]
    if "총연체금액(미해제)_전년말기준)" in df_11:
        feats["총연체금액"] = df_11["총연체금액(미해제)_전년말기준)"]
    
    # 3) 연체대환 관련
    if "연체대환대출건수(미해지)_전년말기준" in df_11:
        feats["연체대환건수"] = df_11["연체대환대출건수(미해지)_전년말기준"]
    
    # 4) 현재 미해제 연체
    if "연체건수(미해제)(대지급포함)_전년말기준" in df_11:
        feats["미해제연체건수"] = df_11["연체건수(미해제)(대지급포함)_전년말기준"]
    
    return feats
```

```python
"""
개인CB/통신CB 파생변수를 만든 이유와 역할
---------------------------------------

[개인CB 파생변수]
- 소비비율_3Mvs1Y  : 최근 단기 소비 증가율 → 소비 급증 집단의 리스크 검증
- 할부비율_3M      : 단기 부채 의존도 → 잠재 연체 위험 반영 여부 확인
- 카드평균이용액_3M : 카드당 평균 사용 강도 → 고소득/과소비 집단 구분
- 신규카드비율_6M  : 최근 신용 확장 속도 → 과도한 신용 팽창 집단 위험 점검
- 체크카드비율     : 보수적 소비 성향 집단 → 낮은 위험 반영 여부 확인
- 한도소진율       : 한도 대비 사용 비율 → 과소비/여력 부족 집단 확인
- 단기대출비율     : 단기대출 의존도 → 연체 위험 증가 신호 검증
- 최초개설_경과일   : 신용 히스토리 길이 → 짧은 신용이력 집단의 위험 확인
- 카드개설_경과일   : 전체 카드 보유 경력 → 신용 안정성 검증

[통신CB 파생변수]
- 연체율_1M / 3M / 6M / 1Y : 기간별 연체 발생률 → 단기/장기 납부습관 확인
- 최장연체일수             : 연체 심각도 지표 → 위험 집단의 지속성 검증
- 연체대환건수             : 대환대출로 막은 연체 → 숨겨진 리스크 검증
- 미해제연체건수           : 현재 진행 중인 연체 → 실질적 위험 수준 확인

[종합적 역할]
- 내부 모델(M1)의 출력값: 연체 위험도, 신용등급 변화
- 외부 CB 지표의 역할   : 집단 단위 소비·신용행태·연체습관을 통해
                          내부 모델이 특정 집단을 과대/과소평가했는지 검증
- 즉, 개인CB는 '소비·신용여력·이력' 관점, 
     통신CB는 '납부성실·연체패턴' 관점에서
     모델 출력 결과를 외부적으로 보정하는 참고 지표로 사용된다.
"""
```
