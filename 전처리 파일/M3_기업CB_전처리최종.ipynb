### 선정 컬럼 결측치 처리

```python
import pandas as pd
import numpy as np

def fill_missing_values_separate(df, num_features_zero=[], num_features_median=[], cat_features=[]):
    """
    원본 df를 복사하여 결측치 처리 후, 
    결측치 여부(_was_null, _missing)는 별도 DataFrame으로 분리하여 반환.
    
    Returns
    -------
    df_filled : 결측치 처리 완료된 DataFrame
    df_missing_flags : 결측치 여부 컬럼만 따로 모은 DataFrame
    """

    df = df.copy()
    missing_flags = pd.DataFrame(index=df.index)

    # 의미 있는 0 처리
    for col in num_features_zero:
        if col in df.columns:
            flag_col = f"{col}_was_null"
            missing_flags[flag_col] = df[col].isna().astype(int)
            df[col] = df[col].fillna(0)

    # 중앙값 처리
    for col in num_features_median:
        if col in df.columns:
            flag_col = f"{col}_was_null"
            missing_flags[flag_col] = df[col].isna().astype(int)
            median_val = df[col].median()
            df[col] = df[col].fillna(median_val)

    # 범주형 처리
    for col in cat_features:
        if col in df.columns:
            flag_col = f"{col}_missing"
            missing_flags[flag_col] = df[col].isna().astype(int)
            df[col] = df[col].fillna("Unknown")

    # 반환
    return df, missing_flags
```

```python
num_features_zero = [
    # 재무제표 중 “값이 없으면 실제 0일 가능성이 큰 항목”
    "현금", "현금등가물", "당좌자산", "재고자산", "유형자산",
    "단기차입금", "차입금", "자본잉여금", "이익잉여금",
    "매출액", "전기매출액", "매출원가", "매출총이익", 
    "판매비와관리비", "법인세비용차감전 순이익", "전기법인세차감전순이익"
]

num_features_median = [
    # 자산/부채/자본 주요 항목
    "유동자산", "비유동자산", "자산총계",
    "유동부채", "부채총계", "자본총계",

    # 날짜형은 중앙값 처리 대신 그대로 둠 (결측 가능)
    "설립일자", "상장일자", "상장폐지일자",

    # 시점 기준
    "기준년월"
]

cat_features = ["업종(중분류)", "외감구분"]

company_202108_filled, company_2021_missing = fill_missing_values_separate(
    df=company_202108_selected,
    num_features_zero=num_features_zero,
    num_features_median=num_features_median,
    cat_features=cat_features
)

test_df_filled, test_df_missing = fill_missing_values_separate(
    df=test_df_selected,
    num_features_zero=num_features_zero,
    num_features_median=num_features_median,
    cat_features=cat_features
)
test_df_selected
```

### 파생 지표 생성

```python
import pandas as pd
import numpy as np

def make_derived_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    M3 기업CB 데이터를 기반으로 주요 파생지표 22개 + 결측치 자동 보정 생성.
    (1) 날짜 → 연·월·일 분해
    (2) 업종/주소 결측 처리
    (3) 재무비율 파생 생성
    (4) inf/NaN 자동 처리 (업종별 평균 후 0 대체)
    """

    df = df.copy()

    # ────────────────────────────────
    # ① 날짜형 처리 (int → 문자열 분리)
    # ────────────────────────────────
    date_cols = ["기준년월", "설립일자", "상장일자", "상장폐지일자"]
    for col in date_cols:
        s = df[col].fillna(0).astype(int).astype(str).str.zfill(8)
        df[f"{col}_년"] = s.str[:4].astype(float)
        df[f"{col}_월"] = s.str[4:6].astype(float)
        df[f"{col}_일"] = s.str[6:8].astype(float)

    # ────────────────────────────────
    # ② 상장 여부 처리
    # ────────────────────────────────
    df["상장여부"] = df["상장일자"].notnull().astype(int)
    df["상장폐지여부"] = df["상장폐지일자"].notnull().astype(int)
    df["상장일자"] = df["상장일자"].fillna(0).astype(int)
    df["상장폐지일자"] = df["상장폐지일자"].fillna(0).astype(int)

    # ────────────────────────────────
    # ③ 업종·주소 결측 처리
    # ────────────────────────────────
    if "업종(중분류)" in df.columns:
        df["업종(중분류)_missing"] = df["업종(중분류)"].isna().astype(int)
        df["업종(중분류)"] = df["업종(중분류)"].fillna("Unknown")

    if "주소지시군구" in df.columns:
        df["주소지시군구_missing"] = df["주소지시군구"].isna().astype(int)
        df["주소지시군구"] = df["주소지시군구"].fillna("Unknown")

    # ────────────────────────────────
    # ④ 기간형 파생
    # ────────────────────────────────
    df["설립연수"] = df["기준년월_년"] - df["설립일자_년"]
    df["상장연수"] = df["기준년월_년"] - df["상장일자_년"]
    df["상장연수"] = df["상장연수"].fillna(0)

    for col in date_cols:
        df[col] = df[col].fillna(0).astype(int)

    # ────────────────────────────────
    # ⑤ 재무비율 파생 생성
    # ────────────────────────────────
    df["유동비율"] = df["유동자산"]*100 / df["유동부채"]
    df["당좌비율"] = (df["당좌자산"] + df["현금등가물"])*100 / df["유동부채"]
    df["현금비율"] = (df["현금"] + df["현금등가물"])*100 / df["유동부채"]

    df["부채비율"] = df["부채총계"]*100 / df["자본총계"]
    df["자기자본비율"] = df["자본총계"]*100 / df["자산총계"]
    df["차입금의존도"] = df["차입금"]*100 / df["자산총계"]
    df["단기차입금비율"] = df["단기차입금"]*100 / df["차입금"]
    df["고정비율"] = df["비유동자산"]*100 / df["자본총계"]
    df["유보율"] = df["이익잉여금"]*100 / df["자본총계"]

    df["매출총이익률"] = df["매출총이익"]*100 / df["매출액"]
    df["영업이익률"] = (df["매출총이익"]*100 - df["판매비와관리비"]) / df["매출액"]
    df["순이익률"] = df["법인세비용차감전 순이익"]*100 / df["매출액"]
    df["ROA"] = df["법인세비용차감전 순이익"]*100 / df["자산총계"]
    df["ROE"] = df["법인세비용차감전 순이익"]*100 / df["자본총계"]

    df["매출성장률"] = (df["매출액"] - df["전기매출액"])*100 / df["전기매출액"]
    df["이익성장률"] = (df["법인세비용차감전 순이익"] - df["전기법인세차감전순이익"])*100 / df["전기법인세차감전순이익"]

    df["총자산회전율"] = df["매출액"]*100 / df["자산총계"]
    df["자본회전율"] = df["매출액"]*100 / df["자본총계"]

    # ────────────────────────────────
    # ⑥ 결측치 및 무한값 처리
    # ────────────────────────────────
    ratio_cols = [
        "유동비율", "당좌비율", "현금비율",
        "부채비율", "자기자본비율", "차입금의존도",
        "단기차입금비율", "고정비율", "유보율",
        "매출총이익률", "영업이익률", "순이익률",
        "ROA", "ROE", "총자산회전율", "자본회전율"
    ]

    # (1) inf/-inf → NaN
    df[ratio_cols] = df[ratio_cols].replace([np.inf, -np.inf], np.nan)

    # (2) 업종별 평균 대체
    if "업종(중분류)" in df.columns:
        df[ratio_cols] = (
            df.groupby("업종(중분류)")[ratio_cols]
              .transform(lambda x: x.fillna(x.mean()))
        )

    # (3) 남은 NaN → 0 (분모=0 or 데이터 누락)
    df[ratio_cols] = df[ratio_cols].fillna(0)

    return df
```

```python
def encode_industry_region(df: pd.DataFrame) -> pd.DataFrame:
    """
    업종(중분류) 및 주소지시군구를 모델 입력용 코드로 변환.
    - Unknown은 항상 0으로 인코딩
    - 나머지는 1부터 시작
    - 이미 존재하는 *_missing 플래그는 그대로 유지
    """
    df = df.copy()

    # ────────────────────────────────
    # 업종(중분류)
    # ────────────────────────────────
    if "업종(중분류)" in df.columns:
        df["업종(중분류)"] = df["업종(중분류)"].fillna("Unknown").astype(str)
        # 대분류 추출
        df["업종대분류"] = df["업종(중분류)"].str[0].where(df["업종(중분류)"] != "Unknown", "Unknown")

        # 업종(중분류) 인코딩 (Unknown=0)
        cat_map = {cat: idx + 1 for idx, cat in enumerate(sorted(df["업종(중분류)"].dropna().unique())) if cat.lower() != "unknown"}
        df["업종(중분류)_code"] = df["업종(중분류)"].map(cat_map).fillna(0).astype(int)

        # 업종대분류 인코딩 (Unknown=0)
        cat_map_main = {cat: idx + 1 for idx, cat in enumerate(sorted(df["업종대분류"].dropna().unique())) if cat.lower() != "unknown"}
        df["업종대분류_code"] = df["업종대분류"].map(cat_map_main).fillna(0).astype(int)
        print("\n[업종대분류 인코딩 매핑 결과]")
        for k, v in cat_map_main.items():
            print(f"   '{k}'  →  {v}")
        print(f"   'Unknown'  →  0")

    # ────────────────────────────────
    # 주소지시군구
    # ────────────────────────────────
    if "주소지시군구" in df.columns:
        df["주소지시군구"] = df["주소지시군구"].fillna("Unknown").astype(str)

        # 시도코드 추출
        def extract_sido(x):
            try:
                x = str(int(float(x)))  # 11680.0 → "11680"
                return int(x[:2])       # "11"
            except:
                return 0

        df["시도코드"] = df["주소지시군구"].apply(extract_sido).astype(int)

        # 주소지시군구 인코딩 (Unknown=0)
        cat_map_region = {cat: idx + 1 for idx, cat in enumerate(sorted(df["주소지시군구"].dropna().unique())) if cat.lower() != "unknown"}
        df["주소지시군구_code"] = df["주소지시군구"].map(cat_map_region).fillna(0).astype(int)

        # 시도코드 인코딩 (Unknown=0)
        cat_map_sido = {code: idx + 1 for idx, code in enumerate(sorted(df["시도코드"].dropna().unique())) if code != 0}
        df["시도코드_code"] = df["시도코드"].map(cat_map_sido).fillna(0).astype(int)

        print("\n[시도코드 인코딩 매핑 결과]")
        for k, v in cat_map_sido.items():
            print(f"   {k}  →  {v}")
        print(f"   0 (Unknown)  →  0")

    return df
```
